WordNet to find the semantic similarity between words and sentences in the document.(word disambiguation)-
From text corpus, find all the synsets of the words from sentences. (perform tokenization, stopword removal & POS tagging for preprocessing)

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.corpus import genesis
nltk.download('genesis')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
text = genesis.raw('english-web.txt')
print(text)
```

```python
tokens = nltk.word_tokenize(text)
print(tokens)
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
filtered_words = [token for token in tokens if token.lower() not in stopwords]
print(filtered_words)
```

```python
pos_tags = nltk.pos_tag(filtered_words)
print(pos_tags)
```

```python
tags = ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ','JJ', 'JJR', 'JJS']
sorted_pos_tag_words = []
for i in range(len(pos_tags)):
  if pos_tags[i][1] in tags:
    sorted_pos_tag_words.append(pos_tags[i])

sorted_pos_tag_words
```

```python
first_10_words = sorted_pos_tag_words[:10]
first_10_words
```

```python
wn.synsets('beginning')
```

```python
wn.synset('begin.v.02').lemma_names()
```

```python
sys = []
for i in range(len(first_10_words)):
    sys.append(wn.synsets(first_10_words[i][0]))

similar_words = []
for i in range(len(sys)):
    for j in range(len(sys[i])):
        similar_words.append(wn.synset(sys[i][j].name()).lemma_names())


set_list = []

for i in range(len(similar_words)):
    for j in range(len(similar_words[i])):
        if(similar_words[i][j] not in set_list):
            set_list.append(similar_words[i][j])
set_list
```
