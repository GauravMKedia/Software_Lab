```python
import nltk
from nltk.corpus import genesis
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Load the Genesis corpus
nltk.download('genesis')
nltk.download('stopwords')
genesis_text = genesis.raw('english-kjv.txt')
```

```python
# Tokenize the text into a list of words
word_list = word_tokenize(genesis_text)
```

```python
# Remove punctuation and stopwords from the list of words
punctuations = ['.', ',', ';', ':', '(', ')', '[', ']', '{', '}', '!', '?', '-', '—', '“', '”', '‘', '’', '...', '...']
stop_words = set(stopwords.words('english'))
```

```python
filtered_words = [word.lower() for word in word_list if word.lower() not in stop_words and word not in punctuations]
```

```python
# Perform stemming on the remaining words
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]
```

```python
# Perform POS tagging on the stemmed words
tagged_words = nltk.pos_tag(stemmed_words)
```
