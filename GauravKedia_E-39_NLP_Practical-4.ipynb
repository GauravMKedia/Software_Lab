```python
raw_data1= state_union.raw('2002-GWBush.txt')
raw_data1
```

```python
from nltk import pos_tag
from nltk import word_tokenize
from nltk import RegexpParser
nltk.download('punkt')
```

```python
list_of_words=word_tokenize(raw_data1)
list_of_words
```

```python
nltk.download('averaged_perceptron_tagger')
tagged_word=pos_tag(list_of_words)
tagged_word 
```

```python
#creating the empty list to store noun,adjectives, verbs
nouns = []
adjectives = []
verbs = []
```

```python
for word, tag in tagged_word:
    if tag.startswith('N'):
        nouns.append(word)
    elif tag.startswith('J'):
        adjectives.append(word)
    elif tag.startswith('V'):
        verbs.append(word)
```

```python
print("Nouns: ", nouns[:25])
print("\nAdjectives: ", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

# PorterStemmer()
 It is based on a set of heuristics that remove suffixes from words to obtain their root or stem. PorterStemmer is known for its simplicity and speed, but it may not always produce accurate stems.

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

for word, tag in tagged_word:
    stemmed_word = stemmer.stem(word)
    if tag.startswith('N'):
        nouns.append(stemmed_word)
    elif tag.startswith('J'):
        adjectives.append(stemmed_word)
    elif tag.startswith('V'):
        verbs.append(stemmed_word)
```

```python
print("Nouns:", nouns[:25])
print("\nAdjectives:", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

# SnowballStemmer()
It is a more advanced version that supports stemming in multiple languages. SnowballStemmer is implemented in many programming languages and provides improved accuracy compared to the original PorterStemmer.

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')

for word, tag in tagged_word:
    stemmed_word = stemmer.stem(word)
    if tag.startswith('N'):
        nouns.append(stemmed_word)
    elif tag.startswith('J'):
        adjectives.append(stemmed_word)
    elif tag.startswith('V'):
        verbs.append(stemmed_word)
```

```python
print("Nouns:", nouns[:25])
print("\nAdjectives:", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

# LancasterStemmer()

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()

for word, tag in tagged_word:
    stemmed_word = stemmer.stem(word)
    if tag.startswith('N'):
        nouns.append(stemmed_word)
    elif tag.startswith('J'):
        adjectives.append(stemmed_word)
    elif tag.startswith('V'):
        verbs.append(stemmed_word)
```

```python
print("Nouns:", nouns[:25])
print("\nAdjectives:", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

# RegexpStemmer()

```python
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer('ing$|s$|ed$', min=4)

for word, tag in tagged_word:
    stemmed_word = stemmer.stem(word)
    if tag.startswith('N'):
        nouns.append(stemmed_word)
    elif tag.startswith('J'):
        adjectives.append(stemmed_word)
    elif tag.startswith('V'):
        verbs.append(stemmed_word)
```

```python
print("Nouns:", nouns[:25])
print("\nAdjectives:", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

# WordNetLemmatizer()

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for word, tag in tagged_word:
    if tag.startswith('N'):
        nouns.append(lemmatizer.lemmatize(word, pos='n'))
    elif tag.startswith('J'):
        adjectives.append(lemmatizer.lemmatize(word, pos='a'))
    elif tag.startswith('V'):
        verbs.append(lemmatizer.lemmatize(word, pos='v'))
```

```python
print("Nouns:", nouns[:25])
print("\nAdjectives:", adjectives[:25])
print("\nVerbs:", verbs[:25])
```

```python
from spacy.symbols import VERB, NOUN, ADJ, ADV

nlp = spacy.load('en_core_web_sm')
base_word = "play"

morph_rules = {
    VERB: ["s", "ed", "ing"],
    NOUN: ["s", "er", "ers", "ing"],
    ADJ: ["er", "est", "ly"],
    ADV: ["ly"]
}

print("Output: ")
for pos in morph_rules:
    for rule in morph_rules[pos]:
        surface_form = base_word + rule
        doc = nlp(surface_form)
        for token in doc:
            if token.pos == pos:
                print(surface_form, token.lemma_, token.pos_)
```
