```python
import spacy
```

```python
text= "Marie Curie, née Maria Sklodowska, was born in Warsaw on November 7, 1867, the daughter of a secondary-school teacher. She received a general education in local schools and some scientific training from her father. She became involved in a students’ revolutionary organization and found it prudent to leave Warsaw, then in the part of Poland dominated by Russia, for Cracow, which at that time was under Austrian rule. In 1891, she went to Paris to continue her studies at the Sorbonne where she obtained Licenciateships in Physics and the Mathematical Sciences. She met Pierre Curie, Professor in the School of Physics in 1894 and in the following year they were married. She succeeded her husband as Head of the Physics Laboratory at the Sorbonne, gained her Doctor of Science degree in 1903, and following the tragic death of Pierre Curie in 1906, she took his place as Professor of General Physics in the Faculty of Sciences, the first time a woman had held this position. She was also appointed Director of the Curie Laboratory in the Radium Institute of the University of Paris, founded in 1914."
```

```python
nlp = spacy.load('en_core_web_sm')
nlp_text=nlp(text)
[(i, i.label_, i.label) for i in nlp_text.ents]
```

```python
from spacy import displacy
doc = nlp(text)

displacy.render(nlp_text, style='ent', jupyter=True)
```

```python
from spacy import displacy
doc = nlp('Marie Curie was a trailblazer in the fields of physics and chemistry, as well as an advocate for gender equality')
displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```

```python
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
tuples = []
for ent in doc.ents:
    for ancestor in ent.root.ancestors:
        if ancestor.ent_type_:
            tuples.append((ancestor.ent_type_, ent.text, ancestor.dep_))

for tuple in tuples:
    print(tuple)

print("Total No. of tuples extracted are:", len(tuples))
```
